## Les différents types d'IA

![Carte de l'IA et du Deep Learning](https://upload.wikimedia.org/wikipedia/commons/thumb/7/71/Carto_IA_deepLearning.svg/langfr-2560px-Carto_IA_deepLearning.svg.png)

L’intelligence artificielle (IA) regroupe différentes approches permettant aux systèmes informatiques de résoudre des problèmes, d’apprendre et de prendre des décisions.

### Apprentissage Automatique (Machine Learning)

L’apprentissage automatique permet aux systèmes d’améliorer leurs performances à partir de données. Il comprend plusieurs sous-catégories :
- Apprentissage supervisé : le modèle apprend à partir de données étiquetées (ex. classification d’images ou reconnaissance de texte).
- Apprentissage non supervisé : il identifie des motifs et des structures dans des données non étiquetées (ex. segmentation de clients, réduction de dimensionnalité).
- Apprentissage semi-supervisé : combine des données étiquetées et non étiquetées pour améliorer les performances du modèle.
- Apprentissage par renforcement : le système apprend par essais et erreurs en interagissant avec son environnement et en recevant des récompenses (ex. agents intelligents dans les jeux vidéo, robots autonomes).

**L’apprentissage profond (Deep Learning)**

Une sous-catégorie du Machine Learning qui utilise des réseaux de neurones artificiels profonds pour traiter des données complexes comme des images, du texte ou du son. Il repose sur des architectures avancées, telles que :
- Les réseaux de neurones convolutifs (CNN) pour l’analyse d’images
- Les réseaux récurrents (RNN, LSTM) pour le traitement du langage naturel
- Les transformeurs (ex. GPT, BERT) pour les modèles de langage modernes

### Systèmes Experts

Les systèmes experts sont des programmes reposant sur une base de connaissances et des règles logiques pour simuler le raisonnement humain dans un domaine précis. Ils sont principalement utilisés pour :
- Le diagnostic médical (ex. systèmes d’aide au diagnostic)
- L’analyse financière (ex. détection de fraudes)
- La planification industrielle (ex. assistance à la maintenance prédictive)

Contrairement aux approches basées sur l’apprentissage automatique, ces systèmes ne s’adaptent pas aux nouvelles données de manière autonome et nécessitent des mises à jour manuelles.

### Recherche et Optimisation Algorithmique

Ce domaine de l’IA concerne la résolution de problèmes complexes à l’aide de méthodes mathématiques et algorithmiques. Il inclut :
- Les algorithmes de recherche heuristique (ex. A*, Dijkstra) pour la planification et la navigation
- Les algorithmes évolutionnaires (ex. algorithmes génétiques, optimisation par essaims particulaires) pour la recherche de solutions optimales
- Les techniques d’optimisation combinatoire (ex. programmation linéaire, recuit simulé) utilisées en logistique, en finance ou en conception de réseaux

## Apprentissage Supervisé vs Non Supervisé

L'apprentissage supervisé et l'apprentissage non supervisé sont deux approches majeures du **Machine Learning**, adaptées à des objectifs et des types de données différents.

![Apprentissage Supervisé Vs Non Supervisé](https://upload.wikimedia.org/wikipedia/commons/thumb/4/40/Apprentissage_Supervisé_Vs_Non_Supervisé.png/620px-Apprentissage_Supervisé_Vs_Non_Supervisé.png)

### 1. Apprentissage Supervisé

**Définition :**  
L'apprentissage supervisé consiste à entraîner un modèle à partir de **données étiquetées**, où chaque entrée est associée à une sortie connue.

**Principe :**  
Le modèle apprend à **associer les entrées aux sorties** en minimisant l'erreur entre ses prédictions et les réponses correctes.

**Exemples :**
- **Classification** : Prédire des catégories (ex. reconnaissance de chiffres manuscrits).
- **Régression** : Prédire une valeur numérique (ex. estimation du prix d'un bien immobilier).

**Avantages :**
- Haute précision si les données sont bien annotées.
- Approprié pour des problèmes où les résultats sont connus.

**Inconvénients :**
- Requiert un grand volume de **données étiquetées**, coûteuses à produire.
- Moins efficace pour des tâches exploratoires.

**Exemple Visuel (gauche de l'image) :**
- Chaque point est associé à un **label connu** (rouge ou bleu).
- Le modèle apprend à **séparer les deux classes** pour prédire les nouvelles observations.

---

### 2. Apprentissage Non Supervisé

**Définition :**  
L'apprentissage non supervisé analyse des **données non étiquetées** et cherche à découvrir des **patterns cachés**.

**Principe :**  
Le modèle explore les structures internes des données **sans indication préalable** et tente d’organiser ou réduire ces données.

**Exemples :**
- **Clustering (regroupement)** : Segmenter des clients selon leur comportement.
- **Réduction de dimensionnalité** : Simplifier des jeux de données complexes (ex. compression d’images).

**Avantages :**
- Fonctionne sans avoir besoin de **données étiquetées**.
- Utile pour explorer des données inconnues.

**Inconvénients :**
- Résultats parfois plus **difficiles à interpréter**.
- Pas de garantie d'obtenir des **groupes significatifs**.

**Exemple Visuel (droite de l'image) :**
- Aucun label n'est donné : l’algorithme découvre **deux groupes distincts** (rouge et bleu).
- Il s'agit d'une **classification automatique** sans intervention humaine.

---

### Résumé Comparatif

| Critère                  | Apprentissage Supervisé          | Apprentissage Non Supervisé         |
|--------------------------|----------------------------------|-------------------------------------|
| **Données**              | Étiquetées (input + output connus) | Non étiquetées (seulement input)   |
| **Objectif**             | Prédire des résultats connus      | Découvrir des patterns cachés      |
| **Exemples de tâches**   | Classification, régression        | Clustering, réduction de dimension |
| **Précision**            | Élevée (si données bien annotées) | Variable (dépend des patterns)     |
| **Complexité**           | Requiert plus de données étiquetées | Plus adaptable mais moins explicite|
| **Exemples d'algorithmes**| Régression linéaire, SVM, Random Forest | K-means, PCA, DBSCAN               |

## Les quatre familles principales d'algorithmes d'apprentissage

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
"""
Dans cette section, nous allons créer un jeu de données simple pour illustrer le concept de régression linéaire. 
La régression linéaire est utilisée pour prédire une valeur continue en supposant une relation linéaire 
entre les features et le label.
"""

# Génération d'un jeu de données synthétique
np.random.seed(42)
x = np.linspace(0, 10, 50)
y = 2.5 * x + np.random.normal(0, 3, 50)

# Affichage du scatter plot des données
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", label="Données observées")
plt.title("Visualisation des données - Régression Linéaire")
plt.xlabel("Feature (x)")
plt.ylabel("Label (y)")
plt.legend()
plt.grid(True)
plt.show()
```

### Régression
![Régression Linéaire](https://upload.wikimedia.org/wikipedia/commons/thumb/3/3a/Linear_regression.svg/langfr-520px-Linear_regression.svg.png)


```python
"""
Régression Linéaire :

La régression linéaire est un algorithme d'apprentissage supervisé qui vise à établir une relation linéaire 
entre une variable indépendante (feature) et une variable dépendante (label).
On cherche à ajuster une droite (ou un hyperplan en dimensions supérieures) aux données.

Nous allons ajuster un modèle de régression linéaire à notre jeu de données créé précédemment.
"""

# Reshape de x pour correspondre aux exigences de scikit-learn (matrice 2D)
X = x.reshape(-1, 1)
model = LinearRegression()
model.fit(X, y)

# Prédictions du modèle
y_pred = model.predict(X)

# Visualisation de la droite de régression par rapport aux données observées
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color="blue", label="Données observées")
plt.plot(x, y_pred, color="red", linewidth=2, label="Modèle de Régression Linéaire")
plt.title("Régression Linéaire sur le jeu de données")
plt.xlabel("Feature (x)")
plt.ylabel("Label (y)")
plt.legend()
plt.grid(True)
plt.show()

print("Coefficient (pente):", model.coef_[0])
print("Intercept (ordonnée à l'origine):", model.intercept_)
```

#### Coefficient et Intercept dans la Régression Linéaire

Dans le contexte de la régression linéaire, le **coefficient** et l'**intercept** sont des paramètres clés du modèle.

- **Coefficient (pente)** : Il représente l'impact d'une unité de changement dans la variable indépendante (feature) sur la variable dépendante (label). En d'autres termes, c'est la pente de la droite de régression. Un coefficient positif indique que la variable dépendante augmente avec la variable indépendante, tandis qu'un coefficient négatif indique une diminution.

- **Intercept (ordonnée à l'origine)** : Il s'agit de la valeur de la variable dépendante lorsque la variable indépendante est égale à zéro. C'est le point où la droite de régression croise l'axe des ordonnées (y-axis).

Ensemble, ces deux paramètres définissent l'équation de la droite de régression : 

$y = (\text{coefficient} \times x) + \text{intercept}$ 

où $y$ est la variable dépendante,  $x$ est la variable indépendante, le coefficient est la pente de la droite, et l'intercept est la valeur de  $y$ lorsque  $x$ est zéro.

### Classification
![Classification](https://quizmanthon.com/images/classification.jpg)

La classification est une méthode d'apprentissage supervisé utilisée pour attribuer une étiquette ou une catégorie à une observation en fonction de ses caractéristiques. Par exemple, dans le dataset Iris, chaque fleur est classée en trois espèces différentes (setosa, versicolor, virginica) en fonction de la longueur et de la largeur de ses sépales et pétales. 

L'algorithme k-NN (k-Nearest Neighbors) est un exemple de méthode de classification qui attribue une catégorie à une nouvelle observation en fonction des catégories de ses k plus proches voisins dans l'espace des caractéristiques.

#### Différence entre Classification et Régression
![Différence entre Classification et Régression](https://cdn.discuss.boardinfinity.com/original/2X/2/2e12f38a6fbdfd17266cff28a62e033578881005.png)

```python
"""
Classification avec k-Nearest Neighbors (k-NN) :

La classification consiste à attribuer une catégorie à un exemple en fonction des caractéristiques observées.
L'algorithme k-NN est une méthode simple qui classe un nouvel exemple selon les labels de ses voisins 
les plus proches dans l'espace des features.

Nous allons utiliser le dataset Iris, un classique en machine learning, pour illustrer cet algorithme.
"""

# Chargement du dataset Iris
iris = load_iris()
X_iris = iris.data[
    :, :2
]  # On utilise les deux premières features pour faciliter la visualisation
y_iris = iris.target

# Création et entraînement du modèle k-NN
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_iris, y_iris)

# Création d'une grille pour visualiser la frontière de décision
h = 0.02  # Pas de la grille
x_min, x_max = X_iris[:, 0].min() - 1, X_iris[:, 0].max() + 1
y_min, y_max = X_iris[:, 1].min() - 1, X_iris[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Prédiction sur la grille
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Visualisation des frontières de décision et des données
plt.figure(figsize=(8, 5))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
plt.scatter(
    X_iris[:, 0], X_iris[:, 1], c=y_iris, edgecolor="k", s=50, cmap=plt.cm.Paired
)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Classification avec k-NN sur le dataset Iris")
plt.show()
```

### Partitionnement (clustering) de données
![Partitionnement de Données](https://upload.wikimedia.org/wikipedia/commons/thumb/a/ad/Hierarchical_clustering_simple_diagram.svg/langfr-520px-Hierarchical_clustering_simple_diagram.svg.png)

Le clustering, ou partitionnement, est une méthode d'apprentissage non supervisé utilisée pour regrouper des données similaires en clusters ou groupes. Contrairement à la classification, où les étiquettes des classes sont connues à l'avance, le clustering ne nécessite pas de données étiquetées. L'objectif est de découvrir des structures cachées dans les données en regroupant les points de données en fonction de leur similarité.

Dans l'exemple ci-dessus, nous avons utilisé l'algorithme K-Means pour partitionner un jeu de données synthétique en 4 clusters. L'algorithme K-Means fonctionne en initialisant aléatoirement des centres de clusters, puis en assignant chaque point de données au centre le plus proche. Les centres sont ensuite mis à jour en calculant la moyenne des points assignés à chaque cluster. Ce processus est répété jusqu'à ce que les centres de clusters ne changent plus significativement.

Le clustering est particulièrement utile pour explorer des données inconnues, segmenter des clients, détecter des anomalies, et bien d'autres applications où la structure des données n'est pas préalablement connue.

#### Différence entre classification et partitionnement (clustering) :
![Clustering vs Classification](https://blog.bismart.com/hubfs/Imported_Blog_Media/Clustering%26clasification-Animales-Sep-26-2023-09-13-22-7262-AM.gif)

La classification attribue des étiquettes prédéfinies à des données en fonction de caractéristiques observées, tandis que le clustering regroupe des données non étiquetées en clusters basés sur leur similarité.

```python
"""
Clustering avec l'algorithme K-Means :

Le clustering est une méthode d'apprentissage non supervisé utilisée pour regrouper des données similaires. 
K-Means partitionne un jeu de données en K clusters en minimisant la variance intra-cluster.

Nous allons générer un jeu de données synthétique et appliquer l'algorithme K-Means pour identifier des clusters.
"""

# Génération d'un jeu de données avec 300 échantillons répartis en 4 clusters
X_blobs, y_blobs = make_blobs(n_samples=300, centers=4, random_state=42)

# Application de l'algorithme K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X_blobs)

# Visualisation des clusters et des centres
plt.figure(figsize=(8, 5))
plt.scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_kmeans, s=50, cmap="viridis")
centers = kmeans.cluster_centers_
plt.scatter(
    centers[:, 0],
    centers[:, 1],
    c="red",
    s=200,
    alpha=0.75,
    marker="X",
    label="Centres des clusters",
)
plt.title("Clustering avec K-Means")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

### Réduction de dimensions
![Réduction de Dimensions](https://upload.wikimedia.org/wikipedia/commons/9/9c/PCA_Projection_Illustration.gif)

La réduction de dimensions consiste à transformer des données présentant un grand nombre de variables en données présentant moins de variables, tout en conservant au maximum la structure ou l'information essentielle des données initiales. Dans ce contexte, l'image illustrée montre par exemple une projection de données d'un espace à haute dimension vers un espace à deux ou trois dimensions, ce qui facilite l'analyse visuelle et la compréhension des relations entre les données.

En pratique, des méthodes telles que l'Analyse en Composantes Principales (PCA) ou l'Analyse en Composantes Indépendantes (ICA) peuvent être utilisées pour réaliser cette transformation. Ces techniques cherchent à réduire le bruit et la redondance, permettant ainsi d'extraire les caractéristiques principales et de simplifier la visualisation et le traitement des données.


```python
"""
    Réduction de dimensions avec PCA :
    L'analyse en composantes principales (PCA) est une technique de réduction de dimensions qui permet de
    projeter les données dans un espace de dimensions inférieures tout en conservant le maximum de variance.
    Nous allons appliquer PCA sur le dataset Iris pour visualiser les données en 2D.
"""

# Chargement du dataset Iris
iris = load_iris()
X_iris = iris.data  # Utilisation de toutes les features
y_iris = iris.target

# Application de PCA pour réduire les dimensions à 2
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_iris)

# Visualisation des données réduites
plt.figure(figsize=(8, 6))
for target, color, label in zip([0, 1, 2], ['r', 'g', 'b'], iris.target_names):
    plt.scatter(X_pca[y_iris == target, 0], X_pca[y_iris == target, 1], color=color, label=label)
plt.xlabel('Composante Principale 1')
plt.ylabel('Composante Principale 2')
plt.title('PCA sur le dataset Iris')
plt.legend()
plt.grid(True)
plt.show()
```

## L'apprentissage par Renforcement

![Diagramme d'apprentissage par Renforcement](https://upload.wikimedia.org/wikipedia/commons/thumb/c/cd/Reinforcement_learning_diagram_fr.svg/440px-Reinforcement_learning_diagram_fr.svg.png)

L'apprentissage par renforcement (RL) est une branche du machine learning où un agent apprend à prendre des décisions en interagissant avec un environnement. L'agent reçoit des récompenses ou des punitions en fonction des actions qu'il entreprend, et son objectif est de maximiser la somme des récompenses reçues au fil du temps. Contrairement à l'apprentissage supervisé, où l'agent apprend à partir de données étiquetées, le RL repose sur l'exploration et l'exploitation pour découvrir les meilleures stratégies d'action.

## Spécificités du Machine Learning

### L'analyse exploratoire des données (EDA)

L'analyse exploratoire des données (EDA) est une étape cruciale dans le processus de data science. Elle consiste à examiner et à visualiser les données pour en comprendre les principales caractéristiques, détecter les anomalies, identifier les patterns et formuler des hypothèses. 

L'EDA utilise des techniques statistiques et graphiques pour résumer les données et en extraire des informations pertinentes. Les outils couramment utilisés pour l'EDA incluent les histogrammes, les diagrammes de dispersion, les matrices de corrélation et les statistiques descriptives.

La méthode **Grid Search** (recherche en grille) teste plusieurs combinaisons d'hyperparamètres pour identifier

### Notions Complémentaires en Machine Learning

- **Biais (Bias)** :
    Erreur due aux suppositions simplificatrices dans le modèle. Un biais trop élevé peut conduire à un sous-apprentissage (underfitting).

- **Variance** :
    Sensibilité du modèle aux fluctuations des données d'entraînement. Une variance trop élevée peut conduire à un surapprentissage (overfitting).

- **Trade-off biais-variance** :
    Équilibre à trouver pour obtenir un modèle qui généralise bien : trop de biais ou trop de variance est néfaste.

- **Régularisation** :
    Technique qui ajoute une pénalité sur la complexité du modèle pour éviter le surapprentissage. Les méthodes Ridge et Lasso sont des exemples de régularisation dans la régression linéaire.

- **Feature Engineering** :
    Processus de création, transformation et sélection de features pertinentes à partir des données brutes.

- **Pipeline de Machine Learning** :
    Enchaînement d'étapes (nettoyage, transformation, modélisation) pour automatiser le processus d'entraînement et de prédiction.

- **Data Augmentation** :
    Technique consistant à générer des variantes des données d'entraînement, particulièrement utilisée en vision par ordinateur.

- **Learning Rate (Taux d'apprentissage)** :
    Paramètre qui détermine la vitesse d'ajustement des paramètres lors de l'entraînement, notamment dans les réseaux de neurones.

- **Epoch** :
    Une passe complète sur l'ensemble du jeu de données d'entraînement lors de l'entraînement d'un modèle.

- **Batch Size** :
    Nombre d'échantillons traités avant la mise à jour des paramètres lors de l'entraînement.

### Exemple final : Prédiction d'une tendance avec Régression Linéaire

Cet exemple montre comment utiliser la régression linéaire pour prédire une tendance future à partir de données historiques. Nous générons des données simulant une tendance avec du bruit, puis nous entraînons un modèle pour faire des prédictions.

```python
# Génération de données simulant une tendance
np.random.seed(100)
x_future = np.linspace(0, 20, 100)
y_future = 3.0 * x_future + np.random.normal(0, 5, 100)

# Séparation des données en ensembles d'entraînement et de test

X_train_future, X_test_future, y_train_future, y_test_future = train_test_split(
    x_future.reshape(-1, 1), y_future, test_size=0.2, random_state=100
)

# Entraînement du modèle de régression linéaire
model_future = LinearRegression()
model_future.fit(X_train_future, y_train_future)

# Prédiction sur l'ensemble complet pour visualiser la tendance
y_pred_future = model_future.predict(x_future.reshape(-1, 1))

# Visualisation des données et du modèle prédit
plt.figure(figsize=(10, 6))
plt.scatter(x_future, y_future, color="gray", label="Données simulées")
plt.plot(
    x_future, y_pred_future, color="blue", linewidth=2, label="Prédiction du modèle"
)
plt.title("Prédiction de tendance avec Régression Linéaire")
plt.xlabel("Temps")
plt.ylabel("Valeur")
plt.legend()
plt.grid(True)
plt.show()

print("Coefficient (pente) du modèle :", model_future.coef_[0])
print("Intercept du modèle :", model_future.intercept_)
```
